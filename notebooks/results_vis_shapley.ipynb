```python
import ast

import matplotlib.pyplot as plt
```

```python
model1_path = '../results/results_our_bad_model_shapley.txt'
model2_path = '../results/results_ours_good_model_global_importance_values.txt'
```

```python
with open(model2_path, 'r') as f:
    lines = ''.join(f.readlines())

lines = ast.literal_eval(lines)
```

```python
lines
```

```python
names = [key for key, _ in lines.items()][:15]
values = [value for __, value in lines.items()][:15]
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

ax.barh(names, values, color='dodgerblue', zorder=2)

ax.set_xlabel("Shapley Values")
ax.set_title('Good model Shapley Value Results for 15 Most Important Features')

ax.invert_yaxis()
ax.grid(axis='x', zorder=0)

fig.tight_layout()

fig.savefig("../figures/good_model_shapley.png")
```

```python
with open(model2_path, 'r') as f:
    model2 = ''.join(f.readlines())

model2 = ast.literal_eval(model2)

model2_top_features = model2[:15]

names2 = [tup[0] for tup in model2_top_features]
values2 = [tup[1] for tup in model2_top_features]
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

ax.barh(names2, values2, color='dodgerblue', zorder=2)

ax.set_xlabel("Feature Importance")
ax.set_title('Model 2 Metamorphic Testing Results for 15 Most Important Features')

ax.invert_yaxis()
ax.grid(axis='x', zorder=0)

fig.tight_layout()

fig.savefig("../figures/model2_metamorphic.png")
```
