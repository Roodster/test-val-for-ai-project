```python
import ast

import matplotlib.pyplot as plt
```

```python
model1_path = '../results/results_model_1_metamorphic_testing_column_avg.txt'
model2_path = '../results/results_model_2_metamorphic_testing_column_avg.txt'
```

```python
with open(model1_path, 'r') as f:
    lines = ''.join(f.readlines())

lines = ast.literal_eval(lines)
```

```python
important_features = lines[:15]

names = [tup[0] for tup in important_features]
values = [tup[1] for tup in important_features]

model1_sum = sum([tup[1] for tup in lines])

normalized_values = [x*100 / model1_sum for x in values]
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

ax.barh(names, normalized_values, color='dodgerblue', zorder=2)

for bar in ax.patches:
    y_value, importance = bar.get_y(), bar.get_width()
    ax.text(importance+0.1, y_value+0.58, f"{importance:.1f}%")


ax.set_xlim(0, 25)

ax.set_xlabel("Feature Importance (%)")
ax.set_title('Model 1 Metamorphic Testing Results for 15 Most Important Features')

ax.invert_yaxis()
ax.grid(axis='x', zorder=0)

fig.tight_layout()

fig.savefig("../figures/model1_metamorphic.pdf")
```

```python
with open(model2_path, 'r') as f:
    model2 = ''.join(f.readlines())

model2 = ast.literal_eval(model2)

model2_top_features = model2[:15]

names2 = [tup[0] for tup in model2_top_features]
values2 = [tup[1] for tup in model2_top_features]

model2_sum = sum([tup[1] for tup in model2])

normalized_values_model2 = [x*100 / model2_sum for x in values]
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

ax.barh(names2, normalized_values_model2, color='dodgerblue', zorder=2)

for bar in ax.patches:
    y_value, importance = bar.get_y(), bar.get_width()
    ax.text(importance+0.05, y_value+0.6, f"{importance:.1f}%")


ax.set_xlim(0, 12)

ax.set_xlabel("Feature Importance (%)")
ax.set_title('Model 2 Metamorphic Testing Results for 15 Most Important Features')

ax.invert_yaxis()
ax.grid(axis='x', zorder=0)

fig.tight_layout()

fig.savefig("../figures/model2_metamorphic.pdf")
```
